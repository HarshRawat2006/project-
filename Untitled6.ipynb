```python
!pip install gtts

from gtts import gTTS
from IPython.display import Audio, display

# Convert text to speech with tone + language
def speak(text, tone="neutral", lang="en", filename="response.mp3"):
    # Tone adjustments (add small prefixes to simulate tone)
    if tone == "friendly":
        text = "😊 " + text
    elif tone == "formal":
        text = "📘 " + text
    elif tone == "neutral":
        text = text
    elif tone == "excited":
        text = "🎉 " + text + "!"
    elif tone == "sad":
        text = "😢 " + text
    elif tone == "robotic":
        text = "🤖 " + text
    else:
        text = f"[{tone}] " + text

    # gTTS conversion
    tts = gTTS(text=text, lang=lang, slow=False)
    tts.save(filename)

    # Play audio in Colab
    display(Audio(filename, autoplay=True))

# ---------------- Run Demo ----------------
speak("Hello! How are you doing today?", tone="friendly", lang="en")
speak("मैं ठीक हूँ, धन्यवाद!", tone="formal", lang="hi")
speak("Wow! This is amazing!", tone="excited", lang="en")
speak("मुझे बहुत दुख है।", tone="sad", lang="hi")
speak("System activated. Ready for commands.", tone="robotic", lang="en")
```

```python
import PyPDF2

def extract_text_from_pdf(pdf_file):
    text = ""
    with open(pdf_file, "rb") as f:
        reader = PyPDF2.PdfReader(f)
        for page in reader.pages:
            text += page.extract_text() + " "
    return text.strip()
```

```python
from gtts import gTTS
from IPython.display import Audio, display

def speak(text, tone="neutral", lang="en", filename="pdf_audio.mp3"):
    # Tone adjustment
    if tone == "friendly":
        text = "😊 " + text
    elif tone == "formal":
        text = "📘 " + text
    elif tone == "neutral":
        text = text
    elif tone == "excited":
        text = "🎉 " + text + "!"
    elif tone == "sad":
        text = "😢 " + text
    elif tone == "robotic":
        text = "🤖 " + text

    # Convert to speech
    tts = gTTS(text=text, lang=lang, slow=False)
    tts.save(filename)
    display(Audio(filename, autoplay=True))
```

```python
def pdf_to_speech(pdf_file, tone="neutral", lang="en"):
    text = extract_text_from_pdf(pdf_file)
    if not text:
        print("❌ No text found in the PDF. Make sure it's not image-only.")
        return
    speak(text, tone=tone, lang=lang)
```
